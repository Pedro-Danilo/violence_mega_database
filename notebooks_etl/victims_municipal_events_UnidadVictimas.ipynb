La estandarización de categorizaciones es el proceso de uniformizar y normalizar los valores de las categorías en un conjunto de datos para asegurar la consistencia y evitar discrepancias. Esto es crucial para la calidad y precisión de los análisis

- Borrar espacios en blanco al principio y al final, cambiar a mayúsculas, remover acentos y eliminar signos extraños

```python
import unicodedata

def remove_accents_and_special_chars(input_str):
    # Normalizar la cadena a NFKD
    nfkd_form = unicodedata.normalize('NFKD', input_str)

    # Eliminar acentos
    no_accents = ''.join([c for c in nfkd_form if not unicodedata.combining(c)])

    # Definir caracteres no deseados
    unwanted_chars = [',', ';', '!', '?', '#', '$', '%']

    # Eliminar caracteres no deseados
    cleaned_str = ''.join([c for c in no_accents if c not in unwanted_chars])

    # Remover espacios en blanco al principio y al final, y convertir a mayúsculas
    result = cleaned_str.strip().upper()

    return result
```

```python
# Aplicar la función a todas las columnas categóricas
victims_municipal_events_UnidadVictimas.loc[:, categorical_col] = victims_municipal_events_UnidadVictimas[categorical_col].astype(str)
for col in categorical_col:
    victims_municipal_events_UnidadVictimas.loc[:,col] = victims_municipal_events_UnidadVictimas[col].apply(remove_accents_and_special_chars)
```

Nota: Como se aprecia todos los muninicipios con codigo de 4 digitos corresponden a los departamentos de ANTIOQUIA Y ATLANTICO. Estos departamentos son los unicos cuyo codigo comienza por cero, el cual por alguna causa fue eliminado en el set de datos encontrado en datos.gov.co


```python
# Adicionar un cero a los codigo_dane de 4 dígitos

# Función que agrega un '0' a la izquierda si la longitud del string es 4
def add_zero_if_length_4(codigo):
    if len(codigo) == 4:
        return '0' + codigo
    return codigo

# Aplicar la función a la columna 'codigo'
victims_municipal_events_UnidadVictimas.loc[:,'cod_ciudad_muni'] = victims_municipal_events_UnidadVictimas['cod_ciudad_muni'].apply(add_zero_if_length_4)

# Verificar cuantos codigo_dane de 4 dígitos quedaron
len(victims_municipal_events_UnidadVictimas[victims_municipal_events_UnidadVictimas['cod_ciudad_muni'].str.len() == 7])
```

```python
# Función para comparar listas y mostrar diferencias
def compare_lists(df1_col, df2_col, label1, label2):
    # Extraer listas únicas y normalizar
    list1 = set(df1_col.str.strip().str.upper().unique())
    list2 = set(df2_col.str.strip().str.upper().unique())

    # Encontrar diferencias
    only_in_list1 = list1 - list2
    only_in_list2 = list2 - list1

    # Imprimir resultados
    print(f"{label1} que no están en {label2}:")
    print(only_in_list1)
    # print(f"{label2} que no están en {label1}:")
    # print(only_in_list2)
    # print()
```

  - Comparar nombres de departamentos en los dataframes

```python
# Comparar listas de departamento
compare_lists(victims_municipal_events_UnidadVictimas['estado_depto'], dept_mpios_codes['dept_name'],
              "Departamentos en victims_municipal_events_UnidadVictimas", "Departamentos dept_mpios_codes")
```

  - Solucionar discrepancias en nombres de departamento

```python
# Diccionario de mapeo basado en los resultados de la comparación departamentos
depto_mapping = {'ARCHIP.DE SAN ANDRES PROVIDENCIA Y SANTA CATALINA': 'ARCHIPIELAGO DE SAN ANDRES PROVIDENCIA Y SANTA CATALINA'}

# Reemplazar los nombres incorrectos  el dataframe arrests_for_ilegal_mining_MinDefensa_
victims_municipal_events_UnidadVictimas['estado_depto'] = victims_municipal_events_UnidadVictimas['estado_depto'].replace(depto_mapping)
```
